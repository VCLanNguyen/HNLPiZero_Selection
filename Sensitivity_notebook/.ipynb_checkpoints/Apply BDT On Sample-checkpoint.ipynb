<h1> Select Highest Energy Shower Object In Slice </h1>

```python
df_hnl.columns
```

```python
#Only Keep Higher Energy Shower PFP per slice
#And only keep slice with valid energy
def cutHighestShowerE(df):
    df = df[df['slc_pfp_shower_energy'] == df.groupby(['run','subrun','event','slc_idx'])["slc_pfp_shower_energy"].transform(max)]
    
    df = df[df['slc_pfp_shower_energy'] > 0]
    
    df = df[df['slc_pfp_shower_dedx'] > 0]
    
    df = df[df['slc_pfp_track_ke'] < 1e18]
    return df
```

```python
df_hnl = cutHighestShowerE(df_hnl)
df_nu = cutHighestShowerE(df_nu)
df_cos = cutHighestShowerE(df_cos)
```

```python
plot_slc_var(df_hnl, df_nu, df_cos,
                    true_counts, start_counts, 
                    'slc_pfp_track_ke', 
                    1000,
                    xmin = 0, xmax = 1e18, xnbin = 50,
                    xtitle = 'Primary Shower Energy'
                    )
plt.show()

plot_slc_var(df_hnl, df_nu, df_cos,
                    true_counts, start_counts, 
                    'slc_total_track_E', 
                    1000,
                    xmin = 1000, xmax = 3000, xnbin = 50,
                    xtitle = 'Primary Shower dEdx'
                    )
```

```python
plot_slc_var(df_hnl, df_nu, df_cos,
                    true_counts, start_counts, 
                    'slc_pfp_shower_energy', 
                    1000,
                    xmin = 0, xmax = 3000, xnbin = 50,
                    xtitle = 'Primary Shower Energy'
                    )
plt.show()

plot_slc_var(df_hnl, df_nu, df_cos,
                    true_counts, start_counts, 
                    'slc_pfp_shower_dedx', 
                    1000,
                    xmin = 1000, xmax = 3000, xnbin = 50,
                    xtitle = 'Primary Shower dEdx'
                    )
```
